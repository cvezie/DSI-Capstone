## 1. Initial Plots

#### Response Rate

```python
fig = plt.figure(figsize=(10,5))
plt.hist(no_out_full['Response_Rate'], bins=20)
plt.title('Response Rate', size=16)
plt.xlabel('Percentage of Returned Surveys')
plt.ylabel('Number of Companies in each Bin')
plt.savefig('./figures/outlier-removed-response_rate.png')
```

#### Alone Share

```python
fig = plt.figure(figsize=(10,5))
plt.hist(no_out_full['Alone_Share'], bins=20)
plt.title('Alone Share', size=16)
plt.xlabel('Share of Employees Driving Alone to Work')
plt.ylabel('Number of Companies in each Bin')
plt.savefig('./figures/outlier-removed-alone_share.png')
```

## 2. Numerical Plots
After removing outliers, what do these features look like?

```python
# just a sample for the presentations.
num_columns_for_pres2 = [
    'worksite_DAR_goal',
    'num_employees_at_worksite',
    'GHGforAgg_(Pounds)',
    'Bike_Share'
]
```

```python
# post outlier removal plots
fig = plt.figure(figsize=(15,8))
fig.suptitle('Numerical Features Distributions After Outliers Removed', fontsize=20, y=1)

for i, num_col in enumerate(num_columns_for_pres2):
    fig.add_subplot(2,2,1+i)
    sns.distplot(no_out_full[num_col])
    plt.xlabel(num_col, size=12)
    
    mean_value = no_out_full[num_col].mean()
    plt.axvline(mean_value, c='red')
    
    median_value = no_out_full[num_col].median()
    plt.axvline(median_value, c='black')
    
fig.tight_layout()
fig.subplots_adjust(top=0.95)
plt.savefig('./figures/numerical_subplots_for_pres_no_outliers.png')
```

## 3. Cost of Program Measured Against `Alone_Share`

```python
x = no_out_full['cost_of_financial_incentives_subsidies_paid_to_employees']
y = no_out_full['Alone_Share']
plt.scatter(x,y)
plt.xlabel('Cost of Financial Incentives Paid to Employees')
plt.ylabel('Alone Share')
plt.savefig('./figures/cost_of_program_against_alone_share.png')
```

## 4. Categorical Plots

```python
categorical_plot_columns_2 = [
    'offers_ORCA_business_passport',
    'offers_drive_alone_parking_subsidy',
    'offers_bike_subsidy',
    'offers_flex_employee_start_end_times',
]
```
